## 리샘플링

```python
df.resample('M').mean()
```

```python
df.resample('M', kind='period').mean()
```

```python
ts = pd.Series(np.arange(12), 
               index = pd.date_range('2020-01-01', periods = 12, freq='T'))
ts
```

```python
ts.resample('5min', closed='right').mean()
```

```python
ts.resample('5min', closed='left').mean()
```

```python
df.resample('D').asfreq()
```

```python
df.resample('D').ffill()
```
